```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/alishbahiqbal/Desktop/c2v_pytorch/Code2vec_Pytorch_pipeline/csvs/q1_marks.csv')
df
```

```python
scaled_class = []
for m in df.Marks:
    if m >= 0 and m <= 4:
        scaled_class += [1]
#         print(m,1)
    elif m>4 and m<=9:
        scaled_class += [2]
#         print(m,2)
    elif m>9:
        scaled_class += [3]
#         print(m,3)
```

```python
df = df.drop('Marks', axis=1)
df['Marks'] = scaled_class
df
```

```python
df.to_csv('/Users/alishbahiqbal/Desktop/c2v_pytorch/Code2vec_Pytorch_pipeline/csvs/q1_scaled_marks.csv', index=False)
```
