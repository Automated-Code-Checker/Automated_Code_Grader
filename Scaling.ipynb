```python
scaled_class = []
for m in df.Marks:
    if m >= 1.08 and m <= 2.666:
        scaled_class += [1]
#         print(m,1)
    elif m>2.666 and m<=6.666:
        scaled_class += [2]
#         print(m,2)
    elif m>6.666:
        scaled_class += [3]
#         print(m,3)
```

```python
df = df.drop('Marks', axis=1)
df['Marks'] = scaled_class
df
```
